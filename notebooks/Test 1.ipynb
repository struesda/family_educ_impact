```python
# Try modeling SEGRADES with logistic regression - using features shown above

#Setup the data with just the columns we will need

focusData = rawData[['SEGRADEQ','GRADEAT','FSSPCOLL','HHUNDR10X','CDOBYY','SESCHWRK','HHUNDR6X','SEENJOY','SEGRADES','SEADPLCX','SEREPEAT','SEFUTUREX','FCSCHOOL']]
focusFeatures = focusData[['SEGRADEQ','GRADEAT','FSSPCOLL','HHUNDR10X','CDOBYY','SESCHWRK','HHUNDR6X','SEENJOY']]
focusTarget = focusData['SEREPEAT']

#Import
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
import scipy.stats as stats

# visualization
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

#Create Objects
logR   = LogisticRegression(C=1e9)

#Split
X_train, X_test, y_train, y_test = train_test_split(focusFeatures,focusTarget,test_size=0.2,random_state=1234)

#Train
logR.fit(X_train, y_train)

#Predict
pred    = logR.predict(X_test)
predProb = logR.predict_proba(X_test)

```

```python
#Calucate accuracy for each of the targets
print("Model accuracy:",metrics.accuracy_score(y_test,pred))

#Get the coefficients - see if we can graph these for each target
logR.coef_

print(list(zip(logR.coef_[0],focusFeatures)))
```
